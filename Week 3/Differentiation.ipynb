### Derivatives

Derivatives are about slopes

So far, we know that slope is a change in y over change in x

$$
\text {Slope} = \frac{\text {change in y}}{\text {change in x}}
$$

So, what's a derivative?

A derivative describes the instanteneous rate of change of a function at a certain point. It gives us the slope of the line tangent to the function's graph at that point .

### Calculating a derivative

Let's use the slope formula to calculate the slope of a fuction $f(x)$

$$
\text {Slope} = \frac{\text {change in y}}{\text {change in x}} = \frac{ y}{ x}
$$

From this diagram:

image.png

We can see that:
1. $x$ changes from $x$ to $x + x$
2. $y$ changes from $f(x)$ to $f(x + x)$

Using the diagram, we can re-write the slope formula as:
$$
\text {Slope} = \frac{ y}{ x} = \frac{f(x + x) - f(x)}{x}
$$

### Actually calculating a derivative

Let's calculate the derivative of this function $f(x) = x^2$ using our re-written slope formula

$$
f(x) = x^2: \frac{(x + x)^2 - x^2}{x}
$$

Epanding $(x + x)^2$ gives us:
$$
= \frac{x^2 + 2xx + (x)^2  x^2}{x}
$$

When we simplify the equation further:
$$
= \frac{2xx + (x)^2}{x}
$$

As $x$ moves towards 0, the slope is:  $2x$

Therefore, the derivative of $x^2$ is $2x$. Or simply put, the slope at $x$ is $2x$.

### Speaking Mathematically...

Instead of writing "as $x$ heads towards 0", mathematicians write $dx$

The derivative, commonly written as $\frac{d}{dx}$

So, we can write the derivative of $x^2$ as:

$$
\frac{d}{dx}x^2 = 2x
$$
