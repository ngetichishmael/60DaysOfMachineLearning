# Support Vector Machine (SVM)

### **Support Vector Machine**

A Support Vector Machine (SVM) is a supervised learning algorithm used for classification and regression tasks. It works by finding the best boundary (hyperplane) that separates different classes in the data. SVM tries to maximize the margin between this boundary and the closest data points (support vectors) from each class. It can handle both linear and non-linear data by using kernel functions. SVMs are effective in high-dimensional spaces and are versatile in terms of classification complexity.


image.png

### **3 Main reasons why we use SVM-**

1. **High-Dimensional Effectiveness:** Performs well with many features.
2. **Kernel Versatility:** Handles linear and non-linear data.
3. **Overfitting Robustness:** Focuses on critical points, reducing overfitting.

### **When we should SVM**

1. **High-Dimensional Data:** When the number of features is large relative to the number of samples.
2. **Non-linear Boundaries:** When the decision boundary between classes is not linear.
3. **Small to Medium-Sized Datasets:** When dealing with datasets where overfitting is a concern but computational resources are limited.




### **Hyperplane in SVM**

1. Hyperplane:

A hyperplane is use to separate different classes of data.
For a 2-dimensional space, this is a line, but in higher dimensions, it becomes a hyperplane.
The goal is to find the hyperplane that maximizes the margin between the classes.

2. Maximizing the Margin:

The SVM algorithm seeks the hyperplane that has the largest distance (margin) to the nearest points from any class, which are called support vectors.
This maximized margin helps improve the generalization ability of the classifier.

3. Support Vectors:

These are the data points that are closest to the hyperplane and influence its position and orientation.
The hyperplane is defined based on these support vectors, rather than the entire dataset.

### **Evaluation Metrics in SVM**

1. **Accuracy**: The proportion of correctly classified instances out of the total instances evaluated.

2. **Precision**: The ratio of true positive predictions to the total predicted positives. It measures the accuracy of positive predictions.

3. **Recall (Sensitivity)**: The ratio of true positive predictions to the total actual positives. It measures the ability of the model to identify all positive instances.

4. **F1 Score**: The harmonic mean of precision and recall. It provides a single metric that balances both precision and recall.

5. **Confusion Matrix**: A table that summarizes the number of true positives, true negatives, false positives, and false negatives. It's useful for understanding where the model is making errors.

6. **ROC Curve (Receiver Operating Characteristic Curve)**: A graphical plot that illustrates the performance of a binary classifier as its discrimination threshold is varied. It plots the true positive rate (TPR) against the false positive rate (FPR).

7. **AUC (Area Under the ROC Curve)**: The area under the ROC curve. It provides an aggregate measure of performance across all possible classification thresholds.

8. **Precision-Recall Curve**: A graphical plot that shows the trade-off between precision and recall for different threshold values.

# 3. Model Building

```python
X = pd[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean', 'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se', 'fractal_dimension_se', 'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst', 'fractal_dimension_worst']]
y = pd['diagnosis']
```

```python
print(pd.head())
print(pd.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Train an SVM model
svm = SVC(kernel='linear', C=1.0, random_state=42)
svm.fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_test)
```

# 4. Evaluation metrics

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
