```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import MeanShift
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./dataset/Dry_Bean_Dataset.csv')
```

```python
X = df.drop(columns=['Class'])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
df_sampled = df.sample(n=1000, random_state=42)
X_sampled = df_sampled.drop(columns=['Class'])
X_sampled_scaled = scaler.fit_transform(X_sampled)
```

```python
ms = MeanShift()
ms.fit(X_sampled_scaled)
```

```python
df_sampled['Cluster'] = ms.labels_
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(X_sampled_scaled[:, 4], 
            X_sampled_scaled[:, 5], 
            c=ms.labels_, cmap='viridis', s=10)
plt.title('Mean Shift Clustering of Sampled Dry Bean Dataset')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.colorbar(label='Cluster Label')
plt.show()
```
